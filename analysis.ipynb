```python
# Importing libraries
import pandas as pd
import numpy as np
pd.options.mode.copy_on_write = True
```

```python
# import as str to ignore conflicts, maybe optimize in latter tests to work match the type.
data = pd.read_csv("DENGBR20.csv", dtype=str)
# ID_O = ID ORIGEM
data.insert(0, "ID_O", range(0, len(data)))

# Lavras-MG 313820
lavras = data[data["ID_MUNICIP"] == "313820"]
lavras.head()
```

Extract only the columns: ID_O, TP_NOT, DT_NOTIFIC, DT_SIN_PRI, NU_IDADE_N, CS_SEXO, SOROTIPO

```python
# ID_F = ID FILTER
filters = ["ID_O", "TP_NOT", "DT_NOTIFIC", "DT_SIN_PRI", "NU_IDADE_N", "CS_SEXO", "SOROTIPO"]
filtered_lavras = lavras[filters]
filtered_lavras.insert(0, "ID_F", range(0, len(filtered_lavras)))
```

```python
# TODO: Fix "ID_F" this will help with the i index in the range function bellow
id_month_pair = []
for i in range(len(filtered_lavras)):
    
    id_month_pair.append([i, filtered_lavras["DT_NOTIFIC"].iat[i].split("-")[1]])
for i in range(len(filtered_lavras)):
    filtered_lavras.loc[i, "DT_NOTIFIC"] = id_month_pair[i][1]
```
